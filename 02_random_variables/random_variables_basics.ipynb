### What is a Random Variable?

A random variable is **a number** that represents the result of a random experiment.

Instead of saying:
- “I got heads”
- “I rolled a six”

We use numbers:
- Heads → 1
- Tails → 0
- Dice roll → 1, 2, 3, 4, 5, 6

That number is the random variable.

In simple math language:
A random variable is **a function that maps an outcome to a number**.


#### Real-life random variables examples:

| **Experiment**   | **Random Variable** |
| ------------     | ----------------    |
| Toss a coin      | Number of heads     |
| Roll a dice      | Dice value          |
| Weather          | Rainfall in mm      |
| Exam             | Marks scored        |
| Online store     | Number of orders    |
| Hospital         | Waiting time        |

### Types of Random Variables :

There are two main types:
1. Discrete Random Variable
2. Continuous Random Variable

### Discrete Random Variable : eg (counting)

A discrete random variable:
- Takes countable values
- Values are usually integers
- can list them one by one

#### Examples:
1. Dice roll
Possible values: 1, 2, 3, 4, 5, 6

Countable → Discrete

2. Coin toss (number of heads)
Tails → 0
Heads → 1

Countable → Discrete

3. Students in a class
10, 11, 12, 13...
10.5 students isn’t possible!

Countable → Discrete

4. Online orders per day
0, 1, 2, 3, ...

Countable → Discrete

### More Examples (Data Science Style)

#### Discrete (counting)
- X = number of customers today
- X = number of errors in logs per hour
- X = number of emails received
- X = number of tickets closed today


### Key Properties of Discrete RV
- Discrete = values are countable (we can list them)
  - Examples:
    - Dice: 1..6
    - Number of calls: 0, 1, 2, 3, ...
    - Defects in a product: 0, 1, 2, ...
  Even if it goes to infinity, it’s still countable.

- Values are usually integers
- Often visualized using bars / spikes


```python
# Discrete random variable example

import random

# Roll a dice 10 times
dice_rolls = [random.randint(1, 6) for _ in range(10)]
print(dice_rolls)
```

### Continuous Random Variable: (eg :measuring)

A continuous random variable:
- Takes any value in a range
- Includes decimals
- Cannot be counted one by one

#### Examples:
1. Time taken to run 100m

- 12.1 sec
- 12.15 sec
- 12.153 sec
Infinite possibilities! hence continuous RV

2. Rainfall in a day

- 0.5 mm
- 3.27 mm
- 10.001 mm
Infinite possibilities! hence continuous RV

3. Temperature

- 22.4°C
- 22.45°C
- 22.451°C
Infinite possibilities! hence continuous RV

4. Stock price
- 102.35
- 102.356
- 102.3567
Infinite possibilities! hence continuous RV

#### More Examples (Data Science Style):
#### Continuous (measuring)
- X = temperature today
- X = time to complete a task
- X = delivery time in minutes


### Key Properties of Continuous RV

- Continuous = values are in a range, with decimals. We can’t list them one by one.
- Measured, not counted
- Represented using smooth curves

Examples:
- time: 3.12 seconds, 3.121 seconds, 3.1211 seconds…
- temperature: 22.4°C, 22.41°C, …


### Quick Compare (Discrete vs Continuous)

- Discrete = counting (0, 1, 2, 3, ...)
- Continuous = measuring (any decimal in a range)

If you can count outcomes one-by-one → **Discrete**.
If you measure with decimals → **Continuous**.


### Real-Time Style Examples (Simple)

These are values that can change right now as time passes:

#### Discrete (counting)
- X = number of people currently online on a website
- X = number of orders coming in this minute
- X = number of support tickets opened today

#### Continuous (measuring)
- X = current temperature in a city
- X = current internet speed (Mbps)
- X = time taken for a page to load (seconds)


```python
# Python example for a continuous Random variable

import random

# Random temperature between 20 and 30
temperature = random.uniform(20, 30)
print(temperature)
```
