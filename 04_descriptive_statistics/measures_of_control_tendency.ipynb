## What is Statistics?

Statistics is the science of collecting, organizing, analyzing, and interpreting data.

It helps us:

- Understand patterns
- Summarize large datasets
- Make data-driven decisions
- Support predictions and modeling

Example:

Instead of analyzing 1 million website visits individually, we summarize them using averages, distributions, and charts.


## Types of Statistics

Statistics is broadly divided into two major categories:

1. Descriptive Statistics
2. Inferential Statistics

Understanding the difference is crucial because each serves a different purpose in Data Science.


### Descriptive Statistics — “What does the data say?”

Descriptive statistics summarizes and describes the dataset you already have.

It does NOT make predictions or assumptions beyond the data.

It answers questions like:

- What is the average value?
- What is the middle value?
- How spread out is the data?
- What patterns exist?

In simple words:

Descriptive statistics = Reporting what happened.

#### Example 1 — Website Analytics

Last month’s traffic report shows:

- Average visitors/day = 12,000
- Median session time = 3.5 minutes
- Bounce rate = 42%

This summarizes past data.

No prediction is made.

→ Descriptive Statistics


#### Example 2 — Salary Report

HR computes:

- Mean salary = €58,000
- Median salary = €51,000
- Highest salary = €210,000

This describes employee salary distribution.

→ Descriptive Statistics


#### Example 3 — Exam Results

A teacher reports:

- Class average = 72%
- Pass rate = 85%
- Most common grade = B

Again, only summarizing results.

→ Descriptive Statistics


### Inferential Statistics — “What can we conclude or predict?”

Inferential statistics goes beyond the dataset.

It uses sample data to make conclusions or predictions about a larger population.

Since we don’t analyze the full population, inference always involves probability and uncertainty.

In simple words:

Inferential statistics = Decision making or prediction.


#### Example 1 — Election Polls

A survey of 2,000 voters shows:

52% support Candidate A.

Conclusion:

Candidate A is likely to win the national election.

We inferred population behavior from a sample.

→ Inferential Statistics


#### Example 2 — A/B Testing

Two website buttons are tested:

- Button A conversion = 8%
- Button B conversion = 10%

We ask:

Is B truly better or is it random chance?

Hypothesis testing is used.

→ Inferential Statistics


#### Example 3 — Drug Trials

A drug is tested on 5,000 patients.

Researchers conclude:

The drug is 78% effective for the population.

Sample → Population conclusion.

→ Inferential Statistics


#### Example 4 — Manufacturing Quality

200 products are inspected.

Defect rate found = 1.2%.

Company estimates defect rate for entire production.

→ Inferential Statistics


### Descriptive vs Inferential Statistics

| Feature | Descriptive | Inferential |
|--------|-------------|-------------|
| Purpose | Summarize data | Predict / Conclude |
| Scope | Dataset only | Population |
| Probability used? | No | Yes |
| Uncertainty involved? | No | Yes |
| Example | Average salary | Predict national salary trend |
| Tools | Mean, charts | Hypothesis tests |


### Quick Intuition Trick

If a statement says:

“The data shows…” → Descriptive

“We conclude / predict…” → Inferential


Examples:

Average app usage is 2.3 hours/day. → Descriptive

Users will likely use the app more next year. → Inferential

- Where data is centered

3. Tables
4. Charts

## Measures of Central Tendency

Measures of Central Tendency describe the “center” or typical value of a dataset.

They answer:

What is the representative value?

The three main measures are:

1. Mean → Arithmetic average
2. Median → Middle value
3. Mode → Most frequent value


### Mean (Average)

Mean is calculated as:

Mean = Sum of all values / Number of values

It is useful when data is evenly distributed.

However, it is sensitive to outliers.


```python
# Example: Mean

visitors = [120, 150, 130, 170, 160, 180, 140] # No of visitors of a website for 7 days

mean_visitors = sum(visitors) / len(visitors)

print("Mean visitors (Average no of visitors per day):", mean_visitors)
```

### Median (Middle Value)

Median is the middle value after sorting the data.

Steps:

1. Sort the dataset.
2. Find the center value : 
    a. If the dataset has odd no of values, then median is the center value.
    b. If the dataset has even no of values, then median is the average of 2 center values.

Median is robust to outliers.


```python
# Example : Median
sales = [12, 15, 14, 10, 18, 120, 16] # Total 7 - odd no of values

sorted_sales = sorted(sales)
median = sorted_sales[len(sorted_sales)//2]

print("Sorted data:", sorted_sales)
print("Median:", median)
```

### Mode (Most Frequent Value)

Mode is the value that appears most frequently in the dataset.

It is useful for categorical data.


```python
from collections import Counter

data = [10, 12, 12, 13, 14, 12, 15]

mode = Counter(data).most_common(1)[0][0]

print("Mode:", mode)
```

## Measures of Dispersion

While central tendency finds the center, dispersion measures spread.

It answers:

How far data points are from the center?

Key measures:

- Range
- Variance
- Standard Deviation


### Range

Range = Max − Min

It shows the total spread of the dataset.


```python
data = [120, 150, 130, 170, 160]

data_range = max(data) - min(data)

print("Range:", data_range)
```

### Variance

Measures average squared deviation from the mean.

### Standard Deviation

Square root of variance.

It shows how concentrated or dispersed data is.


```python
import numpy as np

data = [120, 150, 130, 170, 160]

variance = np.var(data)
std_dev = np.std(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

## Tables in Descriptive Statistics

## Charts in Descriptive Statistics

Charts visualize patterns, trends, and distributions.

Common charts:

- Bar chart → Category comparison
- Line chart → Trends over time
- Histogram → Distribution
- Box plot → Spread + outliers


import matplotlib.pyplot as plt
