Starting from very basics and using a simple example.


## 1) Given Dice Rolls

We are given the results of **12 rolls** of a fair die.

Here is the list of outcomes:

python
rolls = [1, 3, 5, 2, 6, 4, 3, 2, 5, 1, 6, 3]



````python
# List of outcomes of a dice rolled 12 times
rolls = [1, 3, 5, 2, 6, 4, 3, 2, 5, 1, 6, 3]
rolls
````

## 2) Count Each Outcome

We count how many times each number appeared.

A quick way is to use Counter from the collections module.

python
from collections import Counter
counts = Counter(rolls)
counts



````python
from collections import Counter

counts = Counter(rolls)
counts
````

## 3) Experimental Probability of Rolling a 3

Experimental probability is:

Probability = (number of favorable outcomes) / (total outcomes)

For rolling a **3**:

python
P(3) = count_of_3 / total_rolls



````python
total_rolls = len(rolls)
count_3 = counts[3]
p_3 = count_3 / total_rolls
p_3
````

## 4) Experimental Probability of Even Numbers

Even numbers on a die are **2, 4, 6**.

We add the counts of 2, 4, and 6, then divide by total rolls.

python
even_count = count(2) + count(4) + count(6)
P(Even) = even_count / total_rolls



````python
even_count = counts[2] + counts[4] + counts[6]
p_even = even_count / total_rolls
p_even
````

## 5) Compare With Theoretical Probability

For a fair die:

- Theoretical P(3) = 1/6 ≈ 0.1667
- Theoretical P(Even) = 3/6 = 0.5

We compare these with our experimental values.


````python
theoretical_p3 = 1/6
theoretical_p_even = 3/6

print('Experimental P(3) :', round(p_3, 2))
print('Theoretical  P(3) :', round(theoretical_p3, 2))
print('Experimental P(Even) :', round(p_even, 2))
print('Theoretical  P(Even) :', round(theoretical_p_even, 2))
````

## 6) Extra Example: Prime Number

Prime numbers on a die are **2, 3, 5**.

So the event is:

```python
Prime = {2, 3, 5}
```

Theoretical probability:

P(Prime) = 3/6 = 0.5

We also compute experimental probability from our 12 rolls.


````python
prime_count = counts[2] + counts[3] + counts[5]
p_prime = prime_count / total_rolls

print('Experimental P(Prime) :', round(p_prime, 2))
print('Theoretical  P(Prime) :', round(3/6, 2))
````

## 7) Extra Example: At Least 5

At least 5 means **5 or 6**.

Event:

```python
At_least_5 = {5, 6}
```

Theoretical probability:

P(At least 5) = 2/6 = 1/3 ≈ 0.33

Now compute the experimental probability.


````python
at_least_5_count = counts[5] + counts[6]
p_at_least_5 = at_least_5_count / total_rolls

print('Experimental P(At least 5) :', round(p_at_least_5, 2))
print('Theoretical  P(At least 5) :', round(2/6, 2))
````

## 8) Small Simulation (More Rolls)

When we roll a die many times, the **experimental probability** should get closer to the **theoretical probability**.

We will simulate 10,000 rolls and compare: 

- P(3)
- P(Even)

This shows how larger sample sizes reduce randomness.


````python
import random
from collections import Counter

# Simulate many dice rolls
sim_rolls = [random.randint(1, 6) for _ in range(10_000)]
sim_counts = Counter(sim_rolls)

sim_total = len(sim_rolls)
sim_p3 = sim_counts[3] / sim_total
sim_even = (sim_counts[2] + sim_counts[4] + sim_counts[6]) / sim_total

print('Simulation rolls:', sim_total)
print('Experimental P(3) :', round(sim_p3, 3))
print('Theoretical  P(3) :', round(1/6, 3))
print('Experimental P(Even) :', round(sim_even, 3))
print('Theoretical  P(Even) :', round(3/6, 3))
````
