## 1) Core Idea
Probability is a number between **0 and 1**.
- `0` means impossible
- `1` means certain

If all outcomes are equally likely:

\[ P(E) = \frac{\text{favorable outcomes}}{\text{total outcomes}} \]

```python
# Direct formula (no custom probability function)
favorable = 3
total = 10
p_example = favorable / total
p_example
```

## 2) Coin Toss Example
Sample space: `S = {H, T}`
Event: getting Head (`E = {H}`)

```python
sample_space_coin = ['H', 'T']
event_head = ['H']
favorable = len(event_head)
total = len(sample_space_coin)
p_head = favorable / total
p_head
```

## 3) Die Example and Complement
Event A: rolling an even number (`{2,4,6}`)

Complement rule:
\[ P(A^c) = 1 - P(A) \]

```python
sample_space_die = [1, 2, 3, 4, 5, 6]
event_even = [2, 4, 6]
favorable = len(event_even)
total = len(sample_space_die)
p_even = favorable / total
p_not_even = 1 - p_even
p_even, p_not_even
```

## 4) Real-World Example: Weather
If forecast says `P(Rain) = 0.30`, then:
- `P(No Rain) = 1 - 0.30 = 0.70`

```python
p_rain = 0.30
p_no_rain = 1 - p_rain
p_rain, p_no_rain
```

## 5) Real-World Example: Manufacturing Quality
Out of 200 items, 8 are defective.

- `P(Defective) = 8/200`
- `P(Not Defective) = 1 - 8/200`

```python
defective = 8
total_items = 200
p_defective = defective / total_items
p_not_defective = 1 - p_defective
p_defective, p_not_defective
```

## 6) Addition Rule (Mutually Exclusive Events)
For one die roll:
- A = getting 1
- B = getting 6

These cannot happen together in one roll, so:
\[ P(A \cup B) = P(A) + P(B) \]

```python
p_one = 1 / 6
p_six = 1 / 6
p_one_or_six = p_one + p_six
p_one_or_six
```

## 7) Basic Conditional Probability Intuition
Given that a card is a face card (J, Q, K), what is probability it is King?

- Face cards = 12
- Kings = 4
- `P(King | Face Card) = 4/12`

```python
kings = 4
face_cards = 12
p_king_given_face = kings / face_cards
p_king_given_face
```

## 8) Experimental vs Theoretical Probability
- Theoretical: from model assumptions
- Experimental: from observed data

Example below: 20 coin tosses

```python
tosses = ['H','T','H','H','T','H','T','T','H','T','H','H','T','H','T','H','H','T','H','T']
heads_count = tosses.count('H')
tails_count = tosses.count('T')
p_head_exp = heads_count / len(tosses)
p_head_theory = 0.5
heads_count, tails_count, p_head_exp, p_head_theory
```

## 9) Quick Recap
1. `P(E) = favorable / total` (equally likely outcomes)
2. `P(not A) = 1 - P(A)`
3. For mutually exclusive A and B: `P(A or B) = P(A) + P(B)`
4. Conditional form `P(A|B)` means A under the condition B is known
5. Experimental values become closer to theoretical values with more trials
